**Below is the data dictionary for this dataset**

1. instant: An index representing the record number in the dataset.

2. dteday: Date of the record in the format YYYY-MM-DD.

3. season: Season of the year (1: spring, 2: summer, 3: fall, 4: winter).

4. yr: Year (0: 2011, 1: 2012).

5. mnth: Month (1 to 12).

6. holiday: Binary variable indicating whether it's a holiday or not (1: holiday, 0: not a holiday).

7. weekday: Day of the week (0: Sunday, 1: Monday, ..., 6: Saturday).

8. workingday: Binary variable indicating whether it's a working day or not (1: working day, 0: not a working day).

9. weathersit: Weather situation (1: Clear, Few clouds, Partly cloudy, Partly cloudy, 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist, 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds, 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog).

10. temp: Normalized temperature in Celsius.

11. atemp: Normalized feeling temperature in Celsius.

12. hum: Normalized humidity.

13. windspeed: Normalized wind speed.

14. casual: Count of casual users.

15. registered: Count of registered users.

16. cnt: Total count of bike rentals (casual + registered).
